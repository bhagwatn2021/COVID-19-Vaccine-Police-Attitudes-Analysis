```python
# Reverse score variables based on the highest score available
# https://www.marsja.se/reverse-scoring-in-python/
def reverse_score(col, high):
    '''Reverse scores on given columns
     df = your data frame,
     high = highest score available
     cols = the columns you want reversed in list form'''
    return high-col
```

```python
# Min-max normalize based on the highest score available
# https://www.codecademy.com/article/normalization
def normalize(col, high,low):
    '''Reverse scores on given columns
     df = your data frame,
     high = highest score available
     cols = the columns you want reversed in list form'''
    return (col-low)/(high-low)
```
